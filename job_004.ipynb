```python
import numpy as np
```

```python
mA = np.array([[1, 2, 1], [-2, 1, 1]])
mB = np.array([[3, 1], [-1, 0], [2, 4]])
mC = np.array([[-2, 2], [1, -1], [1, 2]])

m1 = mA + 2 * mC.transpose()
m2 = mB + 2 * mC

print(m1)
print(m2)
```

```python
mZ = np.zeros((8, 8), dtype=np.int16)
mZ[::2, 1::2] = 1
mZ[1::2, ::2] = 1
print(mZ)
```

```python
v1 = np.arange(0, 1.5, 0.1)
v2 = np.arange(1, 7, 1)

v1 = np.reshape(v1, (5, 3))
v2 = np.reshape(v2, (3, 2))

print(v1)
print(v2)

vM = np.dot(v1, v2)

print(vM)
```

```python
v0 = np.arange(1, 72, 2, dtype=np.int32)
v1 = v0[v0 % 3 == 0]

print(v0)
print(v1)
```
